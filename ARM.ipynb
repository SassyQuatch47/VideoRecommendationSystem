```python
import pandas as pd
```

```python
db_ref = pd.read_csv('tmdb_5000_movies.csv').sample(100)
```

```python
db_ref
```

```python
import ast
db_ref["Genres"] = [ast.literal_eval(i) for i in db_ref["Genres"]]
```

```python
db_ref['Genres']
```

```python
data , name , genres = {} , [] , []
for i in db_ref.index:
  data[db_ref.loc[i, "title"]] = db_ref.loc[i, "Genres"]
  name.append(db_ref.loc[i, "title"])
  genres.append(db_ref.loc[i, "Genres"])
```

```python
genres = set(item for sublist in genres for item in sublist)
```

```python
print(genres)
print(len(genres))
```

```python
print(len(name))
```

```python
for k,v in data.items() :
  for i in v : 
    print(f"{k} => {i}")
```

```python
classification_db:dict = dict()
for i in db_ref.index:
  db_extract = db_ref.loc[i]
  classification_db[str(i)] = db_extract['Genres']
```

```python
classification_db
```

```python
transaction_db:dict = dict()
for i in classification_db:
  transaction_db[i] = { genre: genre in classification_db[i] for genre in genres }
```

```python
transaction_db = pd.DataFrame(transaction_db).transpose()
transaction_db
```

```python
import miners

miner = miners.AssociationRuleMiner(support_threshold="100%", confidence_threshold="0%")
```
